```python
import numpy as numpyInstance
import pandas as pandasInstance
import matplotlib.pyplot as matplotlibInstance
import seaborn as seabornInstance
```

```python
weather = pandasInstance.read_csv('weather.csv')
display(weather.info())
```

```python
weather.describe()
```

```python
seabornInstance.scatterplot(x='Temp9am', y='Humidity9am', data=weather)
```

```python
seabornInstance.jointplot(x='Temp9am', y='Humidity9am', data=weather, kind='hex')
```

```python
seabornInstance.lmplot(x='Temp9am', y='Humidity9am', data=weather)
```

```python
X_features = weather[['Humidity3pm']]
X_features
```

```python
Y_prediction = weather['Temp9am']
Y_prediction
```

```python
weather.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,10))
plt.tight_layout()
sns.distplot(weather['MaxTemp'])
```

```python
Y_prediction = weather[['MinTemp']]
print(Y_prediction)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_features,Y_prediction)
temperaturePrediction=model.predict(X_features)
print(temperaturePrediction)
```

```python
matplotlibInstance.scatter(Y_prediction,temperaturePrediction)
matplotlibInstance.xlabel('MinTemp')
matplotlibInstance.ylabel('MaxTemp')
matplotlibInstance.show()
```

```python
sns.distplot(Y_prediction-temperaturePrediction)
```

```python
from sklearn import metrics
metrics.mean_absolute_error(Y_prediction, temperaturePrediction)
```

```python
metrics.mean_squared_error(Y_prediction, temperaturePrediction)
```
